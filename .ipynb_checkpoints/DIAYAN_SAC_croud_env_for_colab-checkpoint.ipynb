### Generating pictures 

```python
WINDOW_SIZE = 320
NUM_LEVELS = 10
color = ["red", "yellow", "green", "blue"]
images = NUM_LEVELS*NUM_LEVELS
square = np.zeros(3*images)
figure = [[0,0,0]]*(3*images) # [1,0,0]-круг, [0,1,0]-квадрат, [0,0,1]-треугольник

def make_image(draw, square, figure, num_of_image, width, k, l, d, num_of_cell, diff = True):
    coordinate1 = k*width + width/2 - d/2
    coordinate2 = l*width + width/2 - d/2
    random_color = color[np.random.randint(len(color))]
    draw.ellipse((coordinate1,coordinate2, coordinate1 + d, coordinate2 + d), 
                  fill = random_color, outline = random_color)
    i = num_of_image*images  + k*num_of_cell + l
    square[i] = math.pi*(d/2)**2
    figure[i] = [1,0,0]
    fig = ['ellipse', 'rectangle', 'triangle']
    random_fig = 'ellipse'
    
    for c1 in range(num_of_cell):
        for c2 in range(num_of_cell):
            if c1 != k or c2 != l :
                other_d = np.random.randint(low = d + 10, high = width) 
                if diff:
                    random_fig = fig[np.random.randint(len(fig))] 
                if random_fig == 'ellipse':
                    coordinate1 = c1*width + width/2 - other_d/2
                    coordinate2 = c2*width + width/2 - other_d/2
                    random_color = color[np.random.randint(len(color))]
                    draw.ellipse((coordinate1,coordinate2,coordinate1 + other_d,coordinate2 + other_d), 
                                          fill=random_color, outline=random_color)
                    i = num_of_image*images + c1*num_of_cell + c2
                    square[i] = math.pi*(other_d/2)**2
                    figure[i] = [1,0,0]
                
                if random_fig == 'rectangle':
                    coordinate1 = c1*width + (width - other_d)/2
                    coordinate2 = c2*width + (width - other_d)/2
                    draw.rectangle((coordinate1, coordinate2,coordinate1 + other_d,coordinate2 + other_d),
                                    fill=random_color, outline=random_color)
                    i = num_of_image*images + c1*num_of_cell + c2
                    square[i] = (other_d)**2
                    figure[i] = [0,1,0]
                
                if random_fig == 'triangle':
                    h = other_d*np.sqrt(3)/2.0 ## высота

                    coordinate11 = c1*width + width/2
                    coordinate12 = c2*width + (width - h)/2

                    coordinate21 = c1*width + (width - other_d)/2
                    coordinate22 = c2*width + width - (width - h)/2

                    coordinate31 = c1*width + (width - other_d)/2 + other_d
                    coordinate32 = c2*width + width - (width - h)/2       

                    draw.polygon((coordinate11, coordinate12,coordinate21,coordinate22,coordinate31,coordinate32),
                                    fill=random_color, outline=random_color)
                    i = num_of_image*images + c1*num_of_cell + c2
                    square[i] = other_d*h/2
                    figure[i] = [0,0,1]
                
                
def make_other_image(draw1, draw2,square, figure, num_of_image1,num_of_image2, width, d, num_of_cell, diff = True):
    fig = ['ellipse', 'rectangle', 'triangle']
    random_fig1 = 'ellipse'
    random_fig1 = 'ellipse'
    for c1 in range(num_of_cell):
        for c2 in range(num_of_cell):
            if diff:
                random_fig1 = fig[np.random.randint(len(fig))]   
                random_fig2 = fig[np.random.randint(len(fig))]   
                
            # первая картинка            
            random_color = color[np.random.randint(low = 0, high = len(color))]
            other_d = np.random.randint(low = d + 10, high = width-5)
            if random_fig1 == 'ellipse': 
                coordinate1 = c1*width + width/2 - other_d/2
                coordinate2 = c2*width + width/2 - other_d/2
                draw1.ellipse((coordinate1,coordinate2,coordinate1 + other_d,coordinate2 + other_d), 
                                      fill=random_color, outline=random_color)
                i = num_of_image1*images + c1*num_of_cell + c2
                square[i] = math.pi*(other_d/2)**2
                figure[i] = [1,0,0]
                
            if random_fig1 == 'rectangle':
                coordinate1 = c1*width + (width - other_d)/2
                coordinate2 = c2*width + (width - other_d)/2
                draw1.rectangle((coordinate1, coordinate2,coordinate1 + other_d,coordinate2 + other_d),
                                fill=random_color, outline=random_color)
                i = num_of_image1*images + c1*num_of_cell + c2
                square[i] = (other_d)**2
                figure[i] = [0,1,0]
                
            if random_fig1 == 'triangle':
                h = other_d*np.sqrt(3)/2.0 ## высота
                
                coordinate11 = c1*width + width/2
                coordinate12 = c2*width + (width - h)/2
                
                coordinate21 = c1*width + (width - other_d)/2
                coordinate22 = c2*width + width - (width - h)/2
                
                coordinate31 = c1*width + (width - other_d)/2 + other_d
                coordinate32 = c2*width + width - (width - h)/2       
                
                draw1.polygon((coordinate11, coordinate12,coordinate21,coordinate22,coordinate31,coordinate32),
                                fill=random_color, outline=random_color)
                i = num_of_image1*images + c1*num_of_cell + c2
                square[i] = other_d*h/2
                figure[i] = [0,0,1]
                
            #вторая картинка            
            random_color = color[np.random.randint(low = 0, high = len(color))]
            other_d = np.random.randint(low = d + 10, high = width-5) 
            if random_fig2 == 'ellipse':                
                coordinate1 = c1*width + width/2 - other_d/2
                coordinate2 = c2*width + width/2 - other_d/2
                draw2.ellipse((coordinate1,coordinate2,coordinate1 + other_d,coordinate2 + other_d), 
                                      fill=random_color, outline=random_color)
                
                i = num_of_image2*images + c1*num_of_cell + c2
                square[i] = math.pi*(other_d/2)**2
                figure[i] = [1,0,0]
                    
            if random_fig2 == 'rectangle':
                coordinate1 = c1*width + (width - other_d)/2
                coordinate2 = c2*width + (width - other_d)/2
                draw2.rectangle((coordinate1, coordinate2,coordinate1 + other_d,coordinate2 + other_d),
                                fill=random_color, outline=random_color)
                
                i = num_of_image2*images + c1*num_of_cell + c2
                square[i] = (other_d)**2
                figure[i] = [0,1,0]
                    
            if random_fig2 == 'triangle':
                h = other_d*np.sqrt(3)/2.0 ## высота
                
                coordinate11 = c1*width + width/2
                coordinate12 = c2*width + (width - h)/2
                
                coordinate21 = c1*width + (width - other_d)/2
                coordinate22 = c2*width + width - (width - h)/2
                
                coordinate31 = c1*width + (width - other_d)/2 + other_d
                coordinate32 = c2*width + width - (width - h)/2       
                
                draw2.polygon((coordinate11, coordinate12,coordinate21,coordinate22,coordinate31,coordinate32),
                                fill=random_color, outline=random_color)
                
                i = num_of_image2*images + c1*num_of_cell + c2
                square[i] = other_d*h/2
                figure[i] = [0,0,1]      

def generate(num_of_cell = 3, diff = True, path = "test_320"):
    small = []     
    

    for i in range(10): 
        width = int(WINDOW_SIZE/num_of_cell)
        
        square = np.zeros(3*images)
        figure = [[0,0,0]]*(3*images)

        image1 = Image.new("RGB", (WINDOW_SIZE,WINDOW_SIZE), "white")
        image2 = Image.new("RGB", (WINDOW_SIZE,WINDOW_SIZE), "white")
        image3 = Image.new("RGB", (WINDOW_SIZE,WINDOW_SIZE), "white")

        draw1 = ImageDraw.Draw(image1)
        draw2 = ImageDraw.Draw(image2)
        draw3 = ImageDraw.Draw(image3)

        num_of_image = np.random.randint(3) 
        d = np.random.randint(low = 5, high = width - 15)   
        k = np.random.randint(num_of_cell)
        l = np.random.randint(num_of_cell)

        if num_of_image == 0:  
            make_image(draw1,square, figure, num_of_image, width, k, l, d, num_of_cell, diff = True)
            make_other_image(draw2, draw3,square, figure, 1, 2, width, d,num_of_cell, diff = True)
        elif num_of_image == 1:
            make_image(draw2,square, figure,num_of_image, width, k, l, d, num_of_cell, diff = True)
            make_other_image(draw1, draw3,square, figure, 0, 2, width, d, num_of_cell, diff = True)
        else:
            make_image(draw3,square, figure,num_of_image, width, k, l, d, num_of_cell, diff = True)
            make_other_image(draw1, draw2,square, figure,0, 1, width, d, num_of_cell, diff = True)
        small.append(num_of_image)

        try:
            os.mkdir("image" + f"/{path}_{num_of_cell}_#_{i}")
        except OSError:
            print("Файл уже существует")

        image1.save("image" + f"/{path}_{num_of_cell}_#_{i}" + "/image1.png", "PNG")
        image2.save("image" + f"/{path}_{num_of_cell}_#_{i}" + "/image2.png", "PNG")
        image3.save("image" + f"/{path}_{num_of_cell}_#_{i}" + "/image3.png", "PNG")
        

        N_pictures = 3
        H = 0
        img = Image.new("RGB", (WINDOW_SIZE*3+2*H, WINDOW_SIZE), "white")
        positions = [i*(WINDOW_SIZE + H) for i in range(N_pictures)]
        for im, x in zip([image1, image2, image3], positions):
            img.paste(im, (x, 0))
        #img.show()
        img.save("image" + f"/{path}_{num_of_cell}_#_{i}" +"/FINAL.png")
        
        #dataframe.loc[i, f'level_{num_of_cell}'] = [small[i], figure, square]
        np.savez("image" + f"/{path}_{num_of_cell}_#_{i}" +'/data.npz', small = small[i], figure = figure, square=square )

        
        del draw1
        del draw2
        del draw3

    return small
```

### Creation of the environment

```python
os.chdir("image")

NUM_PICTURES = 10
CSV_DIR = "data.csv"


class CroudsorsingEnv:
    def __init__(self, render=False):

        self.render = render
        self.max_steps = NUM_PICTURES
        self.action_space = spaces.Discrete(3)
        self.__read_picture_csv()

    def reset(self):
        self.level, self.steps, self.fails = 0, 0, 0
        return self.__get_picture(level=self.level)

    def step(self, action):
        self.reward = -1
        if self.answer == action:
            self.fails = 0
            self.reward = 1
            self.level += 1
        else:
            self.fails +=1
            self.reward = -1

        if self.fails == 2 and self.level > 1:
            self.level -= 1
        elif self.fails == 3 and self.level > 1:
            self.level = 1

        self.steps += 1
        self.done = bool(self.steps == 10)

        if self.level == 10:
            next_state = self.__get_picture(level=9)
        else:
            next_state = self.__get_picture(self.level)
        self.information = {}
        return next_state, self.reward, self.done, self.information

    def __read_picture_csv(self):
        with open(CSV_DIR) as f:
            self.df = pd.read_csv(f)

    def __get_picture(self, level):
        i = np.random.randint(NUM_PICTURES)
        self.answer = self.df[f'level_{level + 1}'][i]
        state = torch.tensor(np.array(Image.open(f'image_{level + 1}_#_{i}/FINAL.png'))).float()
        state = state.unsqueeze(0).permute(0, 3, 1, 2)
        return state
```

### Buffer

```python
class ReplayBuffer:
    def __init__(self, capacity):
        self.capacity = capacity
        self.memory = deque(maxlen=capacity)
        self.Transition = namedtuple('Transition', ('skills_one_hot', 'states', 'actions',
                                                    'rewards', 'next_states', 'dones'))


    def push(self, *args):
        self.memory.append(self.Transition(*args))

    def sample(self, batch_size):
        batch = self.Transition(*zip(*random.sample(self.memory, batch_size)))
        return self.Transition(skills_one_hot = torch.FloatTensor(batch.skills_one_hot),
                               states=torch.cat(batch.states),
                               actions=torch.FloatTensor(batch.actions),
                               rewards=torch.FloatTensor(batch.rewards).unsqueeze(1),
                               next_states=torch.cat(batch.next_states),
                               dones=torch.FloatTensor(batch.dones).unsqueeze(1))

    def __len__(self):
        return len(self.memory)


def weights_init_(layer):
    if isinstance(layer, nn.Linear):
        torch.nn.init.xavier_uniform_(layer.weight)
        #torch.nn.init.constant_(layer.bias, 0)
```

### Neural nets for Actor, Critic and Discriminator
(PreprocessLayers - convolutional layers for image preprocessing)

```python
class PreprocessLayers(nn.Module):
    def __init__(self):
        super(PreprocessLayers, self).__init__()
        self.inner = (nn.Conv2d(3, 1, kernel_size = 2, stride=2, padding=0),
                       nn.Sigmoid(),
                       nn.Conv2d(1, 1, kernel_size = 16, stride=16, padding=0),
                       nn.Sigmoid())

        self.layers_1 = nn.Sequential(*copy.deepcopy(self.inner))
        self.layers_2 = nn.Sequential(*copy.deepcopy(self.inner))
        self.layers_3 = nn.Sequential(*copy.deepcopy(self.inner))

        self.layers_2.load_state_dict(self.layers_1.state_dict())
        self.layers_3.load_state_dict(self.layers_1.state_dict())

        self.fc_size = 100

    def forward(self, state):
        pict1 = state[:, :, :, :320]
        pict2 = state[:, :, :, 320:640]
        pict3 = state[:, :, :, 640:]
        x1 = self.layers_1(pict1).view(-1, self.fc_size)
        x2 = self.layers_2(pict2).view(-1, self.fc_size)
        x3 = self.layers_3(pict3).view(-1, self.fc_size)
        x = torch.cat([x1, x2, x3], dim=1)
        return x

class Critic(nn.Module):
    def __init__(self, state_dim, num_skills, action_dim, hidden_dim=512, context_dim=25):
        super(Critic, self).__init__()

        self.preprocess = PreprocessLayers()

        self.liner_context = nn.Linear(num_skills, context_dim)

        self.linear1 = nn.Linear(state_dim + context_dim, hidden_dim)
        self.linear2 = nn.Linear(hidden_dim, hidden_dim)
        self.linear3 = nn.Linear(hidden_dim, action_dim)

        self.linear4 = nn.Linear(state_dim + context_dim, hidden_dim)
        self.linear5 = nn.Linear(hidden_dim, hidden_dim)
        self.linear6 = nn.Linear(hidden_dim, action_dim)

        self.apply(weights_init_)

    def forward(self, state, context):
        x = self.preprocess(state)
        context = self.liner_context(context)
        x = torch.cat([x, context], dim=1)

        Q1 = F.relu(self.linear1(x))
        Q1 = F.relu(self.linear2(Q1))
        Q1 = self.linear3(Q1)

        Q2 = F.relu(self.linear4(x))
        Q2 = F.relu(self.linear5(Q2))
        Q2 = self.linear6(Q2)

        return Q1, Q2


class Actor(nn.Module):
    def __init__(self, state_dim, num_skills, action_dim, hidden_dim=256, context_dim=25):
        super(Actor, self).__init__()

        self.preprocess = PreprocessLayers()
        self.liner_context = nn.Linear(num_skills, context_dim)


        self.linear1 = nn.Linear(state_dim + context_dim, hidden_dim)
        self.linear2 = nn.Linear(hidden_dim, hidden_dim)
        self.linear3 = nn.Linear(hidden_dim, action_dim)


        self.apply(weights_init_)

    def forward(self, state, context):
        x = self.preprocess(state)
        context = self.liner_context(context)

        x = torch.cat([x, context], dim=1)

        x = F.relu(self.linear1(x))
        x = F.relu(self.linear2(x))
        probs = F.softmax(self.linear3(x))
        return probs

    def sample(self, state, context):
        
        probs = self.forward(state, context)
        dist = Categorical(probs)
        action = dist.sample()
        z = probs == 0.0
        z = z.float() * 1e-8
        log_probs = torch.log(probs + z)
        return action, probs, log_probs


class Discriminator(nn.Module):
    def __init__(self, state_dim, num_skills, hidden_dim=512):
        super(Discriminator, self).__init__()
        self.preprocess = PreprocessLayers()

        self.fc1 = nn.Linear(state_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        self.fc3 = nn.Linear(hidden_dim, num_skills)

        self.apply(weights_init_)

    def forward(self, state, action):
        x = self.preprocess(state)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

### The implementation of SAC algorithm for discrete environments 

```python
class SAC_Discrete:

    def __init__(self):
        self.args = Parametrs()

        self.actor = Actor(self.args.state_dim,  self.args.num_skills, self.args.action_dim)
        self.critic = Critic(self.args.state_dim, self.args.num_skills, self.args.action_dim)
        self.critic_target = Critic(self.args.state_dim, self.args.num_skills, self.args.action_dim)
        self.critic_target.load_state_dict(self.critic.state_dict())
        self.log_alpha = torch.tensor(np.log(self.args.alpha), requires_grad=True)

        self.actor_optimizer = optim.Adam(self.actor.parameters(), lr=self.args.lr_ac)
        self.critic_optimizer = optim.Adam(self.critic.parameters(), lr=self.args.lr_cr)
        self.log_alpha_optimizer = optim.Adam([self.log_alpha], lr=self.args.lr_alpha)

        self.memory = ReplayBuffer(capacity=self.args.buffer_size)
        self.updates = 0
        self.scores, self.skill_scores = [], []
        # DIAYN
        self.disc = Discriminator(self.args.state_dim, self.args.num_skills, self.args.hidden_size)
        self.disc_target = Discriminator(self.args.state_dim, self.args.num_skills, self.args.hidden_size)
        self.disc_target.load_state_dict(self.disc.state_dict())
        self.disc_optim = optim.Adam(self.disc.parameters(), lr=self.args.lr_dc)

        self.prior_probability_of_skill = 1.0 / self.args.num_skills
        self.skills_stat = [[] for _ in range(self.args.num_skills)]


    @property
    def alpha(self):
        return torch.exp(self.log_alpha)

    def act(self, state, context):
        skill_one_hot = torch.FloatTensor(context).unsqueeze(0)
        with torch.no_grad():
            action, _, _ = self.actor.sample(state, skill_one_hot)
        return action.item()

    def update(self):
        if len(self.memory) < self.args.batch_size:
            return
       
        self.updates += 1
        context, state, action, reward, next_state, done = self.memory.sample(self.args.batch_size)
        action = action.unsqueeze(1)

        with torch.no_grad():
            next_action, next_probs, next_log_prob = self.actor.sample(next_state, context)
            Q1_next, Q2_next = self.critic_target(next_state, context)
            Q_next = next_probs * (torch.min(Q1_next, Q2_next) - self.alpha.detach() * next_log_prob)
            Q_next = Q_next.mean(dim=1).unsqueeze(-1)
            target = reward + (1 - done) * self.args.gamma * Q_next

        Q1, Q2 = self.critic(state, context)
        Q1_loss = F.mse_loss(Q1.gather(1, action.long()), target)
        Q2_loss = F.mse_loss(Q2.gather(1, action.long()), target)

        self.critic_optimizer.zero_grad()
        (Q1_loss + Q2_loss).backward()
        self.critic_optimizer.step()

        _, probs_new, log_prob_new = self.actor.sample(state, context)
        Q1_new, Q2_new = self.critic(state, context)
        Q = torch.min(Q1_new, Q2_new)

        policy_loss = (probs_new * ((self.alpha.detach() * log_prob_new) - Q)).mean()
        self.actor_optimizer.zero_grad()
        policy_loss.backward()
        self.actor_optimizer.step()

        self.update_alpha(probs_new, log_prob_new)
        if self.updates % self.args.upd_rate_crirtic == 0:
            self.soft_update(self.critic_target, self.critic, self.args.tau)
            self.dc_update(state, context, action)

    def soft_update(self, target, source, tau):
        for target_param, param in zip(target.parameters(), source.parameters()):
            target_param.data.copy_(target_param.data * (1.0 - tau) + param.data * tau)

    def dc_update(self, state_batch, context_batch, action_batch):
        probs_skills = self.disc(state_batch, action_batch)
        contexts = torch.argmax(context_batch, dim=1)
        disc_loss = F.cross_entropy(probs_skills, contexts)

        self.disc_optim.zero_grad()
        disc_loss.backward()
        self.disc_optim.step()

        self.soft_update(self.disc_target, self.disc, self.args.tau)

    def update_alpha(self, prob, log_prob):
        loss = (prob.detach() * (self.alpha * (log_prob - self.args.desired_entropy).detach())).mean()
        self.log_alpha_optimizer.zero_grad()
        loss.backward()
        self.log_alpha_optimizer.step()


    def save(self, name):
        torch.save(self.actor.state_dict(), name)
        print('------Model Saved------')

    def sample_skill(self):
        skill_num = np.random.randint(0, high=self.args.num_skills)
        skill_one_hot = np.zeros(self.args.num_skills)
        skill_one_hot[skill_num] = 1.
        return skill_one_hot, skill_num

    def estimate_skill_probs(self, state, skill_one_hot, action):
        skill_one_hot = torch.FloatTensor(skill_one_hot).unsqueeze(0)
        score = self.disc_target(state, action)
        prob = F.softmax(score, dim=1)
        prob = prob * skill_one_hot
        prob, _ = torch.max(prob, dim=1, keepdim=False)
        return prob.item()

    def train(self):

        for episode in range(self.args.max_episodes):
            steps, total_reward, total_pseudo_r, done, actions = 0, 0, 0, False, []
            state = env.reset()
            skill_one_hot, skill_num = self.sample_skill()

            while not done:
                if episode > self.args.exploration_episodes:
                    action = self.act(state, skill_one_hot)
                else:
                    action = env.action_space.sample()
                actions.append(action)
                skill_prob = self.estimate_skill_probs(state, skill_one_hot, action)
                pseudo_reward = np.log(max(skill_prob, 1e-6)) - np.log(self.prior_probability_of_skill)
                next_state, reward, done, _ = env.step(action)
                total_reward += reward
                total_pseudo_r += pseudo_reward
                modified_reward = pseudo_reward
                self.memory.push(skill_one_hot, state, action, modified_reward, next_state, int(done))
                self.update()
                state = next_state
                steps += 1
            self.skills_stat[skill_num].append(total_reward)
            self.scores.append(total_reward)
            self.skill_scores.append(total_pseudo_r)
            print(f'Ep: {episode}, sk: {skill_num} || r: {total_reward:.3f}, m: {np.mean(self.scores):.3f} || sk_r: {total_pseudo_r:.3f}, m: {np.mean(self.skill_scores):.3f} || actions:{actions}')
            
```

### Parametrs for training 

```python
class Parametrs:
    def __init__(self):
        self.state_dim = 300
        self.action_dim = 3

        self.gamma = 0.5
        self.tau = 0.05

        self.buffer_size = 10000
        self.batch_size = 256
        self.upd_rate_crirtic = 1
        self.desired_entropy = -self.action_dim

        self.lr_cr = 1e-3
        self.lr_ac = 1e-3
        self.lr_alpha = 1e-3
        self.lr_dc = 1e-3

        self.alpha = 0.2
        self.max_episodes = 10000
        self.exploration_episodes = 20

        self.num_skills = 5
        self.hidden_size = 256
```

```python
def plot_results(agent):
    plt.figure(figsize=(16, 20))
    ax1 = plt.subplot(421)
    ax2 = plt.subplot(422)
    ax1.plot(np.arange(0, len(agent.scores)), agent.scores)
    ax1.axvline(x=agent.args.exploration_episodes, color='g', linestyle='-.', label=f'exploration: {agent.args.exploration_episodes} episodes')
    ax1.set_title(f'Agent reward')
    ax1.set_xlabel('Episode')
    ax1.set_ylabel('Reward')
    ax1.legend()

    ax2.plot(np.arange(0, len(agent.skill_scores)), agent.skill_scores)
    ax2.set_title(f'Skill reward')
    ax2.set_xlabel('Episode')
    ax2.set_ylabel('Reward')
    ax2.legend()
    plt.show()
```

### Results 

```python
plot_results(agent)
```
